```python
import pandas as pd
import numpy as np
dt={'name':['tom','james',np.nan,'messi','steve',np.nan,'jack'],
   'age':pd.Series([25,26,np.nan,23,np.nan,29,23]),
   'rating':pd.Series([4.23,3.24,3.98,2.56,np.nan,4.6,np.nan])}
df=pd.DataFrame(dt)
df
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```
