```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]

plt.figure(figsize=(5,3))
plt.plot(area,price,linewidth=0.5)
plt.title('Area vs price')
plt.xlabel('Area in sr ft')
plt.ylabel('Price in lakh')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(area,price)
plt.title('Area vs Price')
plt.xlabel('Area in sr ft')
plt.ylabel("Price in lakh")
plt.xlim((400,2000))
plt.ylim((5,50))
plt.grid()
plt.show()
```

```python
#remove unwanted data
df['normalized-losses']=df['normalized-losses'].replace('?',np.nan)
df
```

```python
df['normalized-losses']=df['normalized-losses'].astype(float)
df
```

```python
df['normalized-losses']=df['normalized-losses'].fillna(df['normalized-losses']).mean()
df
```

```python
df['horsepower']=df['horsepower'].replace('?',np.nan)
df
```

```python
df['horsepower']=df['horsepower'].astype(float)
df
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower']).mean()
df
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['width'],bins=20)
plt.title('width frequency distribution')
plt.xlabel("frequency")
plt.ylabel('width')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['width'],bins=20)
# df['width'].hist(bins=10)
plt.title('width frequency distribution')
plt.xlabel("frequency")
plt.ylabel('width')
plt.show()
```

```python
# plot mean and median
width_mean=df['width'].mean()
width_median=df['width'].median()
plt.figure(figsize=(8,6))
plt.hist(df['width'],bins=20)
df['width'].hist(bins=10)
plt.title("width frequency distribution")
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='yellow')
plt.xlabel("frequency")
plt.ylabel("width")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['width'],df['price'],color='yellow',label='first')
plt.scatter(df['width']*2,df['price']*2,color='red',label='second')
plt.scatter(df['width']/2,df['price']/2,color='blue',label='third')
plt.plot(df.index,df["width"]*5,color='red')
plt.title('Width vs price scatter')
# plt.legend(loc='upper right')
# plt.legend(loc='lower right')
plt.legend(ncol=4,labelcolor=['blue','red','green','yellow'],loc='upper right',fontsize=12)
plt.xlabel("price")
plt.ylabel('width')
plt.show()
```
